```python
import random

cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

endofGame = False

print("Welcome to blackjack games type 'y' to play: ")

def blackjack():
    """Blackjack Games"""
    #Set blank list
    player_cards = []
    computer_cards = []

    def random_card():
      card = random.choice(cards)
      return card

    def random_player():
        player_cards.append(random_card())
    
    def random_computer(): 
        computer_cards.append(random_card())
        
    #random the card for computer and player
    for x in range (2):
        random_player()
        random_computer()
        
    
    
    shouldContinue = True
    ## Score calc

    while shouldContinue:
    
        player_score = sum(player_cards)
        computer_score = sum(computer_cards)

        print(f"Your cards are {player_cards} and current score is {player_score}")
        print(f"Computer first card is {computer_cards[0]}")

        if player_score == 21:
            print("You got blackjack")
            shouldContinue = False

        if input("Do you want to get more card? type 'y' for more and 'n' to pass") == 'y':
            random_player()
            shouldContinue = True

        ## Condition
        if player_score == 21 and computer_score == 21:
            print("You lose, computer got Blackjack")
        elif computer_score == player_score:
            print("Tie,")


        if shouldContinue == False:
            print(f"Computer cards are {computer_cards} and the score is {computer_score}")

    if input("Play another game? type 'n' to stop") == "n":
        endofGame = True
        
while not endofGame:
    blackjack()
    print()
    print("==========")
```
